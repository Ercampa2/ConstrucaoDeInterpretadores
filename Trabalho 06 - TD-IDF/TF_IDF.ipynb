```python
import requests
from bs4 import BeautifulSoup
import re
import math
import spacy
npl = spacy.load("en_core_web_sm")
```

```python
words = [[]]
sentences = []
wordCountSentence = [[]]
tf = [[]] 
amountPerSentence = []
idf = []
tfidf = [[]]
```

```python
def addUrl(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.content, 'html.parser') 
  paragraphs = soup.find_all("p")

  for paragraph in paragraphs:
    if paragraph.find("span") == None and paragraph.parent.find("a") != None:
      text = npl(paragraph.get_text())
      for sentence in text.sents:
        sentences.append(str(sentence))
        for word in sentence:
          if word.text not in words[0] and word.dep_ != "punct" and word.dep_ != "dep":
            words[0].append(str(word))

addUrl("https://www.techtarget.com/searchenterpriseai/definition/natural-language-processing-NLP")
addUrl("https://hbr.org/2022/04/the-power-of-natural-language-processing")
addUrl("https://monkeylearn.com/natural-language-processing/")
addUrl("https://www.javatpoint.com/nlp")
addUrl("https://www.sas.com/en_us/insights/analytics/what-is-natural-language-processing-nlp.html")
```

```python
for sentence in range(len(sentences)):
  words.append([])
  wordCountSentence.append(0)
  for wordBagIndex in range(len(words[0])):
    countWord = 0
    for word in str(sentences[sentence]).split():
      if words[0][wordBagIndex] == word:
        countWord += 1
        wordCountSentence[sentence+1] += 1
    words[sentence+1].append(countWord)
```

```python
for x in range(1, len(wordCountSentence)):
  tf.append([])
  for y in range(len(words[x])):
    tf[x].append(0.0)
    if wordCountSentence[x] != 0:
      tf[x][y] = words[x][y] / wordCountSentence[x]
```

```python
for y in range(len(words[0])):
  amountPerSentence.append(0)
  for x in range(len(words) - 1):
    if words[x+1][y] != 0:
      amountPerSentence[y] += 1
```

```python
for i in range(len(words[0])):
  if amountPerSentence[i] == 0:
    amountPerSentence[i]= 1
  idf.append(math.log(len(words)-1 / amountPerSentence[i]))
```

```python
for i in range(1, len(tf)):
  tfidf.append([])
  for j in range(len(tf[i])):
    tfidf[i].append(tf[i][j] * idf[j])
```

```python
for i in range(len(tfidf) - 300):
  print(tfidf[i])
```
